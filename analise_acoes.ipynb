```python
# Instalar yfinance para obter dados das ações
!pip install yfinance
!pip install matplotlib seaborn

# Importando as bibliotecas
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações para visualizações
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
# Definindo as 20 melhores ações brasileiras
acoes = ['PETR4.SA', 'VALE3.SA', 'ITUB4.SA', 'BBDC4.SA', 'ABEV3.SA',
         'BBAS3.SA', 'WEGE3.SA', 'GGBR4.SA', 'RENT3.SA', 'CSNA3.SA',
         'RAIL3.SA', 'NTCO3.SA', 'LREN3.SA', 'MGLU3.SA', 'HAPV3.SA',
         'B3SA3.SA', 'CIEL3.SA', 'VVAR3.SA', 'YDUQ3.SA', 'JBSS3.SA']

# Baixando os dados das ações
dados_acoes = yf.download(acoes, period='1y')['Adj Close']

# Visualizando os primeiros dados
dados_acoes.head()
```

```python
# Calculando a variação percentual
variacao_percentual = dados_acoes.pct_change() * 100

# Exibindo a variação percentual
variacao_percentual.tail()
```

```python
# Plotando a variação de uma ação específica
dados_acoes[['PETR4.SA', 'VALE3.SA']].plot(title='Variação de Preço - PETR4 e VALE3')
plt.ylabel('Preço (R$)')
plt.show()

# Plotando a variação percentual de todas as ações
variacao_percentual.plot(title='Variação Percentual de Preço - 20 Ações')
plt.ylabel('Variação (%)')
plt.show()
```

```python
# Estatísticas descritivas
estatisticas = dados_acoes.describe()
print(estatisticas)
```

```python
# Criando um heatmap de correlação entre as variações das ações
correlacao = variacao_percentual.corr()

# Plotando um heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title('Correlação das Variações Percentuais')
plt.show()
```
